````python
# ویژگی های دسته ایی
categorical_cols = ['Country', 'State', 'Gender', 'Senior Citizen', 'Partner', 'Dependents', 'Phone Service',
 'Multiple Lines', 'Internet Service', 'Online Security', 'Online Backup', 'Device Protection', 'Tech Support',
 'Streaming TV', 'Streaming Movies', 'Contract', 'Paperless Billing', 'Payment Method', 'Churn Label']
````

````python
# TODO: متد زیر را برای نشان دادن بار پلات تکمیل و متغیر های ذکر شده را بررسی کنید.

import math
def plot_categorical_bars(df,
                        categorical_columns,
                        rot=None,  # پارامتر چرخش برای برچسب‌های محور X
                        bar_width=0.5,  # عرض میله‌ها
                        palette='tab10',
                        figsize=(20, 45)):
    plt.figure(figsize=figsize)
    for i,col in enumerate(categorical_columns):
        plt.subplot(math.ceil(len(categorical_columns)/2),2,i+1)
        cmap=plt.get_cmap(palette)
        colors=cmap(np.linspace(0,1,len(df[col])))
        plt.bar(df[col].value_counts().index,df[col].value_counts().values,width=bar_width,color=colors)
        plt.xticks(rotation=rot)
        plt.title(col)
        plt.xlabel(col)
        plt.ylabel('count')
    plt.tight_layout()
    plt.show() 
````

````python
# TODO: بررسی بارپلات
plot_categorical_bars(df, categorical_cols)
````

````python
# TODO: متد زیر را برای نشان دادن KDE Plot تکمیل و متغیر های ذکر شده را بررسی کنید.

def plot_numeric_distributions(df, numeric_columns, target_col='Churn Label',bw_adjust=1):
    """
    رسم نمودارهای KDE برای متغیرهای عددی با تفکیک بر اساس متغیر هدف.
    این تابع توزیع متغیرهای عددی را برای هر مقدار از متغیر هدف (مانند 'Churn Label')
    در یک شبکه ۲×۲ نمایش می‌دهد.
    پارامترها:
    df (DataFrame): دیتافریم شامل داده‌ها
    numeric_cols (list): لیست ستون‌های عددی برای رسم نمودار
    target_col (str): نام ستون هدف برای دسته‌بندی (پیش‌فرض 'Churn Label')
    """
    plt.figure(figsize=(10,10))
    for i,col in enumerate(numeric_columns):
        plt.subplot(math.ceil(len(numeric_columns)),2,i+1)
        sns.kdeplot(data=df,x=col,hue=target_col,bw_adjust=bw_adjust)
        plt.title(f'bw={bw_adjust}')
    plt.tight_layout()
    plt.show()
````

````python
# TODO: تشخیص مشتریانی که از روش‌های پرداخت خودکار استفاده می‌کنند
def add_payment_recurrence(df):
    def is_automatic(str1):
        if str1 in ['Bank transfer (automatic)','Credit card (automatic)']:
            return 1
        else:
            return 0
    df['Recurring Payment']=df['Payment Method'].apply(is_automatic)
    return df

df=add_payment_recurrence(df)
````

````python
# تشخیص مشتریانی که پتانسیل فروش افزایشی دارند
def add_upsell_flag(df):
    cltv_mean=df['CLTV'].mean()

    def is_upsell(row):
        if row['Churn Score']>50 and row['CLTV']>cltv_mean:
            return 1
        else:
            return 0
    
    df['Upsell Potential']=df.apply(is_upsell,axis=1)
    return df

df=add_upsell_flag(df)
# df.head(20)
````

            <ul>
                <li>0 تا 33: گروه <strong>Low</strong> (احتمال ترک پایین)</li>
                <li>33 تا 66: گروه <strong>Medium</strong> (احتمال ترک متوسط)</li>
                <li>66 تا 100: گروه <strong>High</strong> (احتمال ترک بالا)</li>
            </ul>

            این دسته‌بندی به مدل کمک می‌کند تا الگوهای رفتاری مشتریان را بهتر درک کرده و احتمال ترک سرویس را به‌طور دقیق‌تری پیش‌بینی کند. ویژگی جدید به نام <strong>Churn Score Group</strong> در دیتافریم ذخیره می‌شود. (تکنیک های Discretization )

````python
#TODO: دسته‌بندی مشتریان بر اساس Churn Score
def add_churn_score_group(df):
    def find_group(number):
        if number>=0 and number <=33:
            return 'Low'
        elif number>33 and number<=66:
            return 'Medium'
        elif number>66 and number<=100:
            return 'High'
        
    df['Churn Score Group']=df['Churn Score'].apply(find_group)
    return df
df=add_churn_score_group(df)
# df.head(50)
````

````python
# TODO: آماده سازی داده های برای استفاده از ColumnTransformer

cols=['Gender', 'Senior Citizen', 'Partner', 'Dependents','Payment Method']


# for col in cols:
#     le=LabelEncoder()
#     X_train[col]=le.fit_transform(X_train[col])
#     X_test[col]=le.transform(X_test[col])





transformer=ColumnTransformer(transformers=[
    ('ordinal',OrdinalEncoder(),['Churn Score Group', 'Contract']),
    ('onehot',OneHotEncoder(sparse_output=False),cols)
    ],
    remainder='passthrough')

pipeline=Pipeline([
    ('preprocessing',transformer),
    ('scaling',StandardScaler())
])
X_train=pipeline.fit_transform(X_train)
X_test=pipeline.transform(X_test)
````

````python
# TODO: تعریف تابع برای نمایش classification report و confusion matrix

def show_metrics(model,X_test,y_test):
    y_test_pred=model.predict(X_test)
    print(f'confusion matrix : ')
    print(confusion_matrix(y_test,y_test_pred))
    print(f'classification report : ')
    print(classification_report(y_test,y_test_pred))
````
